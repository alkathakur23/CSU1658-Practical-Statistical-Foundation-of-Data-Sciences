# Using the same data set

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://vincentarelbundock.github.io/Rdatasets/csv/AER/TeachingRatings.csv"
df = pd.read_csv(url)

# Display the first few rows
df.head()
```

```python
# Q.1Identify duplicate professors
duplicates = df[df.duplicated(subset='prof', keep=False)]
print(f"Number of duplicated instructors: {duplicates['prof'].nunique()}")
print(f"Total duplicate rows: {len(duplicates)}")

# Display first few duplicate records
duplicates.head()
```

```python
mean_age_all = df['age'].mean()
std_age_all = df['age'].std()

print(f"Average age (all observations): {mean_age_all:.2f}")
print(f"Standard Deviation of age (all observations): {std_age_all:.2f}")
```

```python
# Keep only one observation per professor
df_unique = df.drop_duplicates(subset='prof')
print(f"Number of unique instructors: {len(df_unique)}")
```

```python
mean_age_unique = df_unique['age'].mean()
std_age_unique = df_unique['age'].std()

print(f"Average age (unique instructors): {mean_age_unique:.2f}")
print(f"Standard Deviation (unique instructors): {std_age_unique:.2f}")
```

```python
# Q.2Compute mean evaluation by division
avg_eval_div = df.groupby('division')['eval'].mean().reset_index()

# Plot bar chart
sns.barplot(x='division', y='eval', data=avg_eval_div, palette='cool')
plt.title("Average Teaching Evaluation by Division")
plt.xlabel("Course Division")
plt.ylabel("Average Evaluation Score")
plt.show()
```

```python
#Q.3
plt.scatter(df['age'], df['eval'], color='teal', alpha=0.6)
plt.title("Relationship between Age and Teaching Evaluation")
plt.xlabel("Age")
plt.ylabel("Teaching Evaluation Score")
plt.show()
```

```python
#Q.4
sns.scatterplot(x='age', y='eval', hue='gender', data=df, alpha=0.7)
plt.title("Age vs Evaluation Score by Gender")
plt.xlabel("Age")
plt.ylabel("Teaching Evaluation")
plt.legend(title='Gender')
plt.show()
```

```python
#Q.5
sns.scatterplot(x='age', y='eval', hue='gender', style='tenure', data=df, alpha=0.8)
plt.title("Age vs Evaluation by Gender and Tenure Status")
plt.xlabel("Age")
plt.ylabel("Teaching Evaluation")
plt.legend(title='Gender & Tenure')
plt.show()
```
